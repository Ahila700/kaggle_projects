```python
def pysqldf(q):
    return sqldf(q, globals())
```

```python
pysqldf("""SELECT name, type1, hp FROM poke p1 
        WHERE hp = (SELECT MAX(hp) 
                    FROM poke p2 
                    WHERE p1.type1 = p2.type1
                    AND is_legendary = 0) 
        AND is_legendary = 0
        GROUP BY type1, name""")
```
